```python
def calculate_skewness_kurtosis(data, column_name):
    """
    Calculates and prints the skewness and kurtosis for a specified column in the given DataFrame.

    Parameters:
    - data: DataFrame containing the data.
    - column_name: String, the name of the column for which skewness and kurtosis are to be calculated.

    Skewness measures the asymmetry of the probability distribution of a real-valued random variable about its mean. 
    Positive skewness indicates a distribution with an asymmetric tail extending toward more positive values.
    
    Kurtosis measures the "tailedness" of the probability distribution. A higher kurtosis indicates a distribution with heavier tails, 
    and a lower kurtosis indicates a distribution with lighter tails.
    """
    skewness = data[column_name].skew()
    kurtosis = data[column_name].kurtosis()

    print(f"Skewness for {column_name}:", skewness)
    print(f"Kurtosis for {column_name}:", kurtosis)
```

```python
calculate_skewness_kurtosis(data,'Ladder score')
```

```python
def plot_correlation_heatmap(data):
    # Selecting only the numerical columns
    numerical_data = data.select_dtypes(include=['float64', 'int64'])
    
    # Calculating the correlation matrix
    corr_matrix = numerical_data.corr()
    
    # Generating the heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
    plt.title('Correlation Heatmap')
    plt.show()
    return
```

```python
plot_correlation_heatmap(data)
```

```python
def generate_pie_chart_top_countries(data, top_n=10):
    # Selecting the top N countries based on ladder score mean
    top_countries = data.groupby('Country name')['Ladder score'].mean().nlargest(top_n).index
    top_countries_data = data[data['Country name'].isin(top_countries)]
    
    # Calculating the scores for the top N countries
    top_countries_scores = top_countries_data.groupby('Country name')['Ladder score'].sum()

    # Plotting the pie chart without 'Other' category
    plt.figure(figsize=(8, 8))
    plt.pie(top_countries_scores, labels=top_countries_scores.index, autopct='%1.1f%%', startangle=140)
    plt.title(f'Top {top_n} Countries')

    plt.show()
    return
```

# Elbow Method

```python
#The elbow_method function takes the data and the maximum number of clusters as input.
#It iterates over a range of cluster numbers (from 1 to max_clusters) and fits KMeans clustering models to the data for each cluster number.
#For each model, it computes the inertia (sum of squared distances of samples to their closest cluster center) and stores it in a list.
#Finally, it plots the inertia values against the number of clusters and displays the plot.
#You can use this function with your data to visualize the elbow plot and determine the 
#optimal number of clusters for your KMeans clustering. Once you've identified the optimal number of clusters, you can proceed with KMeans clustering using that number of clusters.


from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

def elbow_method(data, max_clusters=10):
    # Exclude non-numeric columns if any
    numeric_data = data.select_dtypes(include=['float64', 'int64'])  # Selecting only numeric columns

    # Create a pipeline with imputer and KMeans
    pipeline = make_pipeline(SimpleImputer(strategy='mean'), KMeans())

    # Initialize list to store inertia values
    inertia = []

    # Fit KMeans with different number of clusters and calculate inertia
    for k in range(1, max_clusters + 1):
        pipeline.steps[-1] = ('kmeans', KMeans(n_clusters=k, random_state=42))  # Update the number of clusters
        pipeline.fit(numeric_data)
        inertia.append(pipeline.steps[-1][1].inertia_)

    # Plot the inertia values
    plt.plot(range(1, max_clusters + 1), inertia, marker='o')
    plt.title('Elbow Method for Optimal Number of Clusters')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Inertia')
    plt.xticks(range(1, max_clusters + 1))
    plt.show()

# Example usage:
# Assuming your DataFrame is named 'df' and you want to use all numeric features for clustering
# elbow_method(df)

```

```python
# elbow prediction

from sklearn.cluster import KMeans
from sklearn.impute import SimpleImputer

def elbow_method(data, max_clusters=10):
    # Impute missing values with the mean for all columns
    imputer = SimpleImputer(strategy='mean')
    imputed_data = imputer.fit_transform(data)
    
    # Initialize list to store inertia values
    inertia = []

    # Fit KMeans with different number of clusters and calculate inertia
    for k in range(1, max_clusters + 1):
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(imputed_data)
        inertia.append(kmeans.inertia_)

    # Find the optimal number of clusters using the elbow method
    optimal_num_clusters = 1
    for i in range(1, len(inertia) - 1):
        if inertia[i] - inertia[i + 1] > 0.1 * inertia[i]:  # Adjust the threshold as needed
            optimal_num_clusters = i + 1
            break

    # Return the optimal number of clusters
    return optimal_num_clusters

# Example usage:
# Assuming your DataFrame is named 'df'
# Drop non-numeric columns if any
numeric_data = data.select_dtypes(include=['float64', 'int64'])

# Determine the optimal number of clusters using the elbow method
optimal_clusters = elbow_method(numeric_data)
print("Optimal number of clusters:", optimal_clusters)

# Fit KMeans with the optimal number of clusters and predict cluster labels
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(numeric_data)

# Display the cluster labels
print("Cluster labels:", cluster_labels)


```

# K-Mean Clustering

```python
from sklearn.cluster import KMeans
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
import seaborn as sns

def perform_clustering(data, n_clusters=3):
    # Selecting the features for clustering
    features = ['Explained by: Log GDP per capita', 'Explained by: Social support',
                'Explained by: Healthy life expectancy', 'Explained by: Freedom to make life choices',
                'Explained by: Generosity', 'Explained by: Perceptions of corruption']

    # Extracting the feature data
    X = data[features]

    # Handling missing values with SimpleImputer
    imputer = SimpleImputer(strategy='mean')
    X_imputed = imputer.fit_transform(X)

    # Performing KMeans clustering
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    data['Cluster'] = kmeans.fit_predict(X_imputed)

    # Visualizing the clusters
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='Explained by: Log GDP per capita', y='Explained by: Social support',
                    hue='Cluster', data=data, palette='viridis', legend='full')

    plt.title('Clustering of Countries based on Happiness Features')
    plt.xlabel('Explained by: Log GDP per capita')
    plt.ylabel('Explained by: Social support')
    plt.show()

    return data


```

```python
clustered_data = perform_clustering(data, n_clusters=3)
```

# Fitting

```python
#Fit classification or regression models to predict
#outcomes based on the cluster labels obtained from clustering.

#We first use the SimpleImputer to handle missing values in the features by replacing them with the mean value of each feature.
#We then fit the linear regression model on the imputed data.
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer

def fit_regression_on_clusters(data):
    # Handling missing values
    imputer = SimpleImputer(strategy='mean')
    X = imputer.fit_transform(data[['Explained by: Log GDP per capita', 'Explained by: Social support']])
    y = data['Ladder score']

    # Fitting linear regression model
    model = LinearRegression()
    model.fit(X, y)

    return model
```

```python

regression_model = fit_regression_on_clusters(clustered_data)
#We select the features 'Explained by: Log GDP per capita' and 'Explained by: Social support' as predictors (X) and the target 'Ladder score' as the outcome variable (y).
#We fit a linear regression model on the clustered data.
#The fit_regression_on_clusters function returns the fitted regression model.
```
